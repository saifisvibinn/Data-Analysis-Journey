```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
import seaborn as sn
```

```python
league = pd.read_csv(r".\Master_Ranked_Games.csv")
```

```python
league['win'] = league['blueWins']
league = league.drop(columns=['blueWins', 'redWins'])
```

```python
y = league['win']
```

```python
X = league.drop(columns=['win'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
model = LogisticRegression(max_iter=10000)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
print("Precision:", metrics.precision_score(y_test, y_pred, average='weighted'))
print("Recall:", metrics.recall_score(y_test, y_pred, average='weighted'))
print("F1 Score:", metrics.f1_score(y_test, y_pred, average='weighted'))
```

```python
print("prediction value : " , y_pred)
```

```python
print("Actual value: ", y_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
decision_tree_model = DecisionTreeClassifier()
decision_tree_model.fit(X_train, y_train)
```

```python
decision_tree_predictions = decision_tree_model.predict(X_test)
```

```python
decision_tree_accuracy = accuracy_score(y_test, decision_tree_predictions)
print(f"Decision Tree Accuracy: {decision_tree_accuracy}")
```
