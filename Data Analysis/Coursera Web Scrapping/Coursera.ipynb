```python
with open("Coursera.csv", "w", encoding='utf-8') as f:
    header = "Course, From, Skill\n"
    f.write(header)

    for page_num in range(1, 84):
        url = f"https://www.coursera.org/search?query=python&page={page_num}&sortBy=BEST_MATCH&=null"

        response = requests.get(url)
        soup = bs(response.content, 'html.parser')

        containers = soup.find_all('div', {'class': 'css-1evtm7z'})

        for container in containers:
            from_tag = container.find_all("p", {"class": "css-vac8rf"})
            from_text = from_tag[0].text.strip() if from_tag else "From not found"

            course_tag = container.find_all("h3", {"class": "cds-CommonCard-title css-6ecy9b"})
            course_name = course_tag[0].text.strip() if course_tag else "Course not found"

            skills_container = container.find("p", {"class": "css-vac8rf"})
            skills_text = ""
            if skills_container:
                skills_div = skills_container.find_next('div', {'class': 'cds-CommonCard-bodyContent'})
                if skills_div:
                    skills_text = skills_div.text.strip().replace("Skills you'll gain: ", "")
                    
                   # print(skills_text)
                   # print(from_text)
                   # print(course_name)

            f.write(f"{course_name.replace(',', ' ').replace('\u3000', ' ')}, {from_text.replace(',', ' ').replace('\u3000', ' ')}, {skills_text.replace(',', ' ').replace('\u3000', ' ')}\n")
```
